```python
# Crear promedios móviles para los últimos 5 juegos como equipo local
df_games['PTS_home_avg_last_5'] = (
    df_games.groupby('HOME_TEAM_ID')['PTS_home']
    .rolling(window=5, min_periods=1)
    .mean()
    .reset_index(0, drop=True)
)

# Crear promedios móviles para los últimos 5 juegos como equipo visitante
df_games['PTS_away_avg_last_5'] = (
    df_games.groupby('VISITOR_TEAM_ID')['PTS_away']
    .rolling(window=5, min_periods=1)
    .mean()
    .reset_index(0, drop=True)
)
```

```python
df_games = df_games[['FG_PCT_home', 'FT_PCT_home', 'FG3_PCT_home', 'AST_home', 'REB_home',
       'TEAM_ID_away', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away',
       'AST_away', 'REB_away', 'HOME_TEAM_WINS', 'PTS_home_avg_last_5', 'PTS_away_avg_last_5' ,'PTS_home', 'PTS_away'
       ]]
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.linear_model import LinearRegression, Lasso, ElasticNet, BayesianRidge
from sklearn.kernel_ridge import KernelRidge
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, make_scorer, r2_score
```

```python
# Dividir en características y variables objetivo
X = df_games.drop(columns=['PTS_home', 'PTS_away'])
y = df_games[['PTS_home', 'PTS_away']]

# Configurar los modelos y sus hiperparámetros
models = {
    'LinearRegression': (LinearRegression(), {}),
    'Lasso': (Lasso(), {'alpha': [0.01, 0.1, 1, 10]}),
    'ElasticNet': (ElasticNet(), {'alpha': [0.01, 0.1, 1], 'l1_ratio': [0.2, 0.5, 0.8]}),
    'KernelRidge': (KernelRidge(), {'alpha': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}),
    'SGDRegressor': (SGDRegressor(), {'alpha': [0.0001, 0.001, 0.01]}),
    'BayesianRidge': (BayesianRidge(), {}),
    'GaussianProcessRegressor': (GaussianProcessRegressor(), {}),
    'RandomForestRegressor': (RandomForestRegressor(), {'n_estimators': [50, 100, 200]}),
    'SVR': (SVR(), {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']})
}

# Crear un diccionario para almacenar resultados
results = []

# Validación cruzada
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scoring = {
    'MAE': make_scorer(mean_absolute_error),
    'MSE': make_scorer(mean_squared_error),
    'R2': make_scorer(r2_score)
}

# Iterar por cada modelo y ajustar con GridSearchCV
for name, (model, params) in models.items():
    grid = GridSearchCV(estimator=model, param_grid=params, scoring=scoring, refit='MAE', cv=kf)
    grid.fit(X, y)
    
    # Obtener los mejores resultados
    best_model = grid.best_estimator_
    mae = grid.cv_results_['mean_test_MAE'][grid.best_index_]
    mse = grid.cv_results_['mean_test_MSE'][grid.best_index_]
    r2 = grid.cv_results_['mean_test_R2'][grid.best_index_]
    
    results.append({
        'Model': name,
        'Best Params': grid.best_params_,
        'MAE': mae,
        'MSE': mse,
        'R2': r2
    })

# Crear un DataFrame con los resultados
results_df = pd.DataFrame(results)
```
